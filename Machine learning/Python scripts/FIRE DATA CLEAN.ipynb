```python
import pandas as pd
import geopandas as gpd
import matplotlib as plt

suomi_viirs_path = "DATASETS/SUOMI_VIIRS_C2_USA_contiguous_and_Hawaii_24h.csv"
noaa_viirs_path = "DATASETS/J1_VIIRS_C2_USA_contiguous_and_Hawaii_24h.csv"
modis_path = "DATASETS/MODIS_C6_USA_contiguous_and_Hawaii_24h.csv"

```

```python
suomi_viirs_data = pd.read_csv(suomi_viirs_path, parse_dates=['acq_date', 'acq_time'])
suomi_viirs_data
```

```python
noaa_viirs_data = pd.read_csv(noaa_viirs_path, parse_dates=['acq_date', 'acq_time'])
noaa_viirs_data
```

```python
modis_data = pd.read_csv(modis_path, parse_dates=['acq_date', 'acq_time'])
modis_data
```

<h2>
    Se eliminan los datos con baja confidencia
</h2>

```python
modis_data = modis_data.drop(modis_data[modis_data.confidence < 90].index)
modis_data
```

```python
suomi_viirs_data = suomi_viirs_data.drop(suomi_viirs_data[suomi_viirs_data.confidence == "low"].index)
suomi_viirs_data
```

```python
noaa_viirs_data = noaa_viirs_data.drop(noaa_viirs_data[noaa_viirs_data.confidence == "low"].index)
noaa_viirs_data
```

<h2>
    Se concatenan los valores de todos los satelites en un solo archivo
</h2>

```python
latitude = pd.concat([modis_data["latitude"], suomi_viirs_data["latitude"], noaa_viirs_data["latitude"]])
longitude = pd.concat([modis_data["longitude"], suomi_viirs_data["longitude"], noaa_viirs_data["longitude"]])
date = pd.concat([modis_data["acq_date"], suomi_viirs_data["acq_date"], noaa_viirs_data["acq_date"]])
hour = pd.concat([modis_data["acq_time"], suomi_viirs_data["acq_time"], noaa_viirs_data["acq_time"]])


data = [latitude, longitude, date, hour]
headers = ["latitude", "longitude", "date", "hour"]

fire_data = pd.concat(data, axis=1, keys=headers).reset_index(drop=True)
fire_data
```

```python
fire_data.to_csv('DATASETS/fire_data.csv',index=False)
```
